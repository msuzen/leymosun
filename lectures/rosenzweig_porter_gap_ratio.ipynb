## Summary

The Rozensweig-Porter (RP) ensemble [rosen] provides one-parameter family of random matrix 
ensemble that one can induce "defects" in the system by injecting different 
structure into energy matrix, i.e., Hamiltonian. This defects or different 
parametrization can identify the range of regimes a physical system can attain 
from Poisson to Wigner-Dyson spectral spacing (nearest-neigbour distance 
distribution).


## RP ensemble 

The random matrix ensemble is defined by addition of two components, 

$H_{rp} = H_{0} + N^{-\gamma/2.0} H_{goe}$.

$H_{goe}$ is drawn from Gaussian Orthogonal Ensemble (GOE). We parametrized 
this component with $\gamma$ that identifies the localisation strength in 
the quantum system. This is related to Anderson localisation. The first 
component is fixed diagonal matrix, i.e., off-diagonal elements are zero, 

$H_{0} \in \mathbb{R}^{N \times N}$.

Diagonal elements of the first component are drawn from a normal
distribution $diag(H_{0}) \in \mathcal{N}(0,1.0)$. The total matrix
is basically computed by element-wise matrix addition. When, $\gamma=0.0$,
$H_{rp}$ is identical to GOE, hence represents quantum chaotic system at that
regime. 

## Different regimes 

Depending on the values of $\gamma$ [pino], the resulting random matrix
represent different phases.

* Wigner-Dyson (Ergodic-Chaotic): $1.0 \gt \gamma \ge 0.0$. 
* Wigner-Dyson-short (Fractal): $2.0 \gt \gamma \gt 1.0$. 
* Poisson (Localized-Regular):  $\gamma \ge 2.0$. 

## Gap ratio 

The gap ratio [huse] statistics measures average consecutive spectral gap
ratio of a the given matrix ensemble, i.e, here RP at a fixed $\gamma$. 
 $\langle r \rangle = \frac{ min(\delta_{n}-\delta{n+1}) }{ max(\delta_{n}-\delta{n+1}) }$
 where eigenenergy difference $\delta=E_{n+1}-E_{n}$. This is also called spectral gap 
 ratio.  Its value is  $\langle r \rangle≈ 0.5307$ for level-spacing distributed as 
 GOE and $\langle r \rangle ≈ 0.3863$ for Poisson [pino]. These corresponds to $\gamma<1.0$
 and $gamma>2.0$ for RP ensemble.  

 In this lecture we will verify these basic definition using `leymosun`'s well 
 tested modules.

```python
from leymosun.rosenzweig import rp
from functools import partial
from leymosun.matrix import ensemble
from leymosun.spectral import empirical_spectral_density, mean_adjacent_gap_ratio
```

```python
def get_rp_mean_gap(gamma):
    rp_sampler = partial(rp, gamma=gamma)
    matrix_order = 500 
    ensemble_size = 20
    ee = ensemble(matrix_order, ensemble_size, rp_sampler)
    eigenvalues, _, _ = empirical_spectral_density(ee, scale='wigner')
    bar_adjacent_gap_ratio, adjacent_gap_ratios = mean_adjacent_gap_ratio(eigenvalues)
    return bar_adjacent_gap_ratio, adjacent_gap_ratios
```

## Chaotic regime

Here we choose $\gamma=0.1$ as a strong chaotic regime for RP and will compute the average 
gap ratio in this case.It must be close to $\langle r \rangle≈ 0.5307$

```python
gamma = 0.1
bar_adjacent_gap_ratio, adjacent_gap_ratios = get_rp_mean_gap(gamma) 
bar_adjacent_gap_ratio # ~0.53
```

## Regular regime

Here we choose $\gamma=6.5$ as a strong regular regime for RP and will compute the average 
gap ratio in this case.It must be close to $\langle r \rangle ≈ 0.3863$

```python
gamma = 6.5
bar_adjacent_gap_ratio, adjacent_gap_ratios = get_rp_mean_gap(gamma) 
bar_adjacent_gap_ratio # ~0.38
```
