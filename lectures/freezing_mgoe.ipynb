## Composite physical system:  A frozen qubit and thermalized system 

Composite systems are quite rich in quantum mechanics [carroll21]. A system 
composed with qubit and thermalized system is proposed [suzen25]. It is written 
as follows:  

In the case of a frozen qubit, $H_{1}=0$ and $H_{int}=0$ . Then,
the composed system, described by the total energy $H$, becomes,

\begin{equation}
H = \mathbb{I}_{2} \otimes H_{2} = \begin{pmatrix}
H_{2} & \bf{0} \\
\bf{0} & H_{2}
\end{pmatrix},
\end{equation}

where $\bf{0} \in \mathbb{C}^{N \times N}$ is the zero matrix.

As $H_{2}$ represents quantum chaotic system or in thermalized
state, due to BGS conjecture \cite{bohigas84}, we can model
$H_{2}$ as a random matrix from Wigner-Dyson family, Gaussian
Orthogonal Ensemble (GOE). Then the spectrum of composite
energy $H$ and its component $H_{2}$ are given by,

$ \sigma(H_{2}) = \{ \lambda_{1},..., \lambda_{N} \}$

$ \sigma(H) = \{ \lambda_{1},..., \lambda{_N}, \lambda_{1}...,\lambda_{N}  \}$.

We conjecture that restricting the spectrum of $H$ by truncating the eigenvalues at a
given fraction—preserving the ordering of the energy levels—we retain the first
$\mu$ percent of the eigenenergies, corresponding to a subsystem of the composite system.
Here, $\mu$ acts as a mixing or effective coupling parameter.

In this notebook we show that inducing periodic boundary conditions (PBC) for   
mGOE (mixed GOE) at a given $\mu$ , is equivalent to partial observation on the  
composite system. 

```python
import numpy as np
import matplotlib.pyplot as plt
import leymosun
from leymosun.gaussian import goe
from leymosun.spectral import apply_pbc
from leymosun.stats import pdf
```

## Basic numerical test

Here, we manually check how eigenvalues are identical and repeating in the   
composite system and individual mixed ensemble matrix. 

```python
matrix_order  = 5
A_goe = goe(matrix_order)
I2 = np.zeros([2,2])
I2[np.diag_indices_from(I2)] = 1.0
Acomposed = np.kron(I2, A_goe)
np.linalg.eigvals(A_goe), np.linalg.eigvals(Acomposed)
```

## Visualizing Wigner-Density with periodicity and truncation

As we mentioned, and detailed in [suzen25], truncation in composite 
physical system. We use `apply_pbc` method from `leymosun` to test this.

```python
matrix_order  = 1000 # wigner Cats with PBC
mixture_order = 800 # Corresponds to mu=0.8
A_goe = goe(mixture_order)
e_i = np.linalg.eigvals(A_goe)
e1 =  apply_pbc(list(e_i), matrix_order)/np.sqrt(mixture_order/ 2.0)
locations: np.array = np.arange(-2.05, 2.1, 0.05)
y, x = pdf(e1, locations)
plt.plot(x, y)
```

```python
I2 = np.zeros([2,2]) # Wigner Cat on composite-full diagonal truncate
I2[np.diag_indices_from(I2)] = 1.0
Acomposed = np.kron(I2, A_goe)
e2 = np.linalg.eigvals(Acomposed)[0:matrix_order]/np.sqrt(mixture_order/ 2.0) 
locations: np.array = np.arange(-2.05, 2.1, 0.05)
y, x = pdf(e2, locations)
plt.plot(x, y)
```
